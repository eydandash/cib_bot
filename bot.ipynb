```python
import os
from pathlib import Path
from processing.parser import process_pdf_document
from embeddings.embed import get_embeddings
from vectorDB.qdrant_scripts import update_vector_store

# Get all PDF files in the statements directory
statements_dir = Path("statements")
pdf_files = list(statements_dir.glob("*.pdf"))

print(f"Found {len(pdf_files)} PDF files to process")

# Initialize collection (recreate only for the first file)
collection_name = "cib_financial_statements"
recreate_collection = True

successful_files = []
failed_files = []

for i, pdf_file in enumerate(pdf_files):
    print(f"\n{'='*60}")
    print(f"Processing {i+1}/{len(pdf_files)}: {pdf_file.name}")
    print(f"{'='*60}")
    
    try:
        # Process the PDF document
        text_pages, image_pages, markdown_sections, combined_results = process_pdf_document(str(pdf_file))
        
        if not combined_results.strip():
            print(f"⚠️  No text extracted from {pdf_file.name}, skipping...")
            failed_files.append(pdf_file.name)
            continue
        
        print(f"✅ Successfully processed: {pdf_file.name}")
        print(f"📄 Pages: {len(text_pages)} text, {len(image_pages)} image")
        print(f"📝 Content: {len(combined_results)} characters")
        
        # Generate embeddings from the parsed text
        texts, embeddings = get_embeddings(combined_results, file_name=pdf_file.name)
        
        # Update the vector store with the new embeddings
        update_vector_store(
            texts=texts,
            embeddings=embeddings,
            collection_name=collection_name,
            host="localhost",
            port=6333,
            recreate_collection=recreate_collection
        )
        
        # Only recreate collection for the first file
        recreate_collection = False
        successful_files.append(pdf_file.name)
        
        print(f"✅ Added {len(texts)} text chunks to vector store")
        
    except Exception as e:
        print(f"❌ Error processing {pdf_file.name}: {e}")
        failed_files.append(pdf_file.name)
        continue

print(f"\n🎯 Vector store '{collection_name}' updated with data from {len(successful_files)} documents")
```
